```python
### params and data source
scan_pfam_path      = '/Users/kszczepaniak/Data/Phage/preliminary-results/hhblits-pfam.txt'
ecf_aho_80_filepath = '/Users/kszczepaniak/Data/Phage/preliminary-results/table-ecf-aho-80'
overlap_threshold   = 0.4 # threshold of hit coverage (how well the Pfam overlaps with hit)

```

```python
# load Pfam scan data (all reprseqs on Pfam HHsuite scan)
df_pfam = pd.read_csv(scan_pfam_path)
df_pfam
```

```python
### add Pfam data to each ECF
# each ECF inherits its source refseq annotation if the Pfam hit sufficiently overlaps with ECF
# overlap is defined as (intersection / union) of ECF span and Pfam hit span
# ECF span is (min,max) of AA positions of all hits comprising it
# Pfam hit span is qstart, qend of the hit in hhr result file

### WARNING: this is desired version of annotation, right now implementation below is different!
```

```python
### this implementation considers all hits comprising ECF and returns sum of all its Pfams

# for each reprseq in cluster_data of each ECF: assign Pfam data
# assign data with some threshold of hit coverage (how well the Pfam overlaps with hit)
cluster_data_update = []
for eid, ecf in table_ecf_80.iterrows():
    seqid   = ecf['prot_id']
    cl_data = ecf['cluster_data']
#     display(cl_data)
    annot = df_pfam[df_pfam.qname == seqid]
#     display(annot)
    cl_data_annot = []
    for hid, hit in cl_data.iterrows():
        hit_span = set(range(hit.qstart, hit.qend+1))
        hits_pfam = []
        for phid, phit in annot.iterrows():
            phit_span = set(range(phit.qstart, phit.qend+1))
            # check if hit overlaps sufficiently with pfam hit
            if (len(hit_span.intersection(phit_span)) / len(hit_span.union(phit_span))) >= overlap_threshold:
                hits_pfam.append(phit.annot)
        cl_data_annot.append(','.join(hits_pfam))
    cl_data = cl_data.assign(annot = cl_data_annot)
    cluster_data_update.append(cl_data)
    
# display(cluster_data_update)
```

```python
table_ecf_80.cluster_data = cluster_data_update
```

```python
table_ecf_80[(table_ecf_80.prot_id == 'reprseq00506') & (table_ecf_80.ecf_id == 1)]['pfams'].to_list()
```

```python
def assign_pfam_to_ecf(ecf):
    
    cl_data = ecf['cluster_data']  
    pfams   = set()
    
    for hid, hit in cl_data.iterrows():
        for pfam in hit.annot.split(','):
            if pfam: # exclude empty string
                pfams.add(pfam)
    
    return ','.join(list(pfams))
```

```python
# propagate Pfam data to ECF table

pfams = []
# apply function that checks cluster_data and returns unique Pfams
for eid, ecf in table_ecf_80.iterrows():
    pfams.append(assign_pfam_to_ecf(ecf))
table_ecf_80['pfams'] = pfams
```
